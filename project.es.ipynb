#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código
calif_estud = [8.5, 9.2, 7.8, 8.9, 10]
print(round(sum(calif_estud)/len(calif_estud), 2))
```

- Encuentra la calificación más alta y la más baja.

```python
# Código
calif_estud = [8.5, 9.2, 7.8, 8.9, 10]
calif_max = calif_estud[0]
calif_min = calif_estud[0]

for x in calif_estud:
    if x > calif_max:
        calif_max = x
    if x < calif_min:
        calif_min = x

print(f"La calificación más alta es {calif_max} y la más baja sería {calif_min}")
```
